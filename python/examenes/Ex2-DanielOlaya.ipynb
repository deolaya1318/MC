```python
#Para comenzar se importa los paquetes requeridos para desarrollar las actividades en este cuaderno
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.fftpack import ifft, fft, fftfreq
from scipy.io import wavfile
from scipy import interpolate
```

## 1.

```python
#Primero se importan adecuadamente los datos al notebook, se toma en cuenta que el tipo de los datos es complejo
#Entonces se usa la herramienta ifft para hallar la señal de audio original. Usando int16 se convierte a enteros
#para luego graficar la señal y guardarla en el archivo ring.wav
dftring = np.genfromtxt('ring.dft', dtype=complex)
f = 22050
N = len(dftring)
t = np.linspace(0,1./f*N,N)
y = np.int16(np.real(ifft(dftring)))
plt.plot(t,y)
wavfile.write("ring.wav",f,y)
```

## 2. a)

```python
#usando un spline de primer orden encontre la función que representa los datos de la señal. Al crear un nuevo array de
#tiempo con el mismo sample rate que el tiempo original, pero con el doble de datos, puedo crear una nueva señal que
#duplique la longitud de la señal a partir de la función hallada 
p = interpolate.interp1d(t,y,kind='slinear')
newt = np.linspace(0,1./f*N,2*N)
newy = p(newt)

#Se grafica la función para verificar que la señal obtenida mantenga la forma de la señal original, pero con el doble
#de datos
plt.plot(newt,newy)
print "Los datos que tiene la señal original son: " + str(len(y))
print "Los datos que tiene la nueva señal son: " + str(len(newy))
```

## b)

```python
#se crea la función rampa usando la función linspace de numpy. Para aplicar la rampa se multiplica esta función a la
#señal hallada mediante la interpolación implementada en el paso anterior. El resultado se grafica y se convierte
#a enteros adecuados para guardar la señal de audio en ring2.wav
rampa = np.linspace(1,0,len(newy))
yrampa = rampa*newy
plt.plot(newt,yrampa)
y2 = np.int16(yrampa)
wavfile.write("ring2.wav",f,y2)
```
