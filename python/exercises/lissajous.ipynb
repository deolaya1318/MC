```python
t = linspace(0,2*pi,100)
A = 1
B = A
a = 3
b = 1
delta = 0
x = A*cos(a*t+delta)
y = B*sin(b*t)
fig=figure()
ax = fig.add_subplot(111, autoscale_on=True)
plot(x,y)
```

```python
fig=figure(figsize=(12,12))
c = 1
for a in range(5):
    for b in range(5):
        delta = 0
        x = A*cos((a+1)*t+delta)
        y = B*sin((b+1)*t)
        subplot(5,5,c)
        axis("off")
        axis("equal")
        plot(x,y)
        c = c+1
fig.savefig('lissajous.png')      
```
