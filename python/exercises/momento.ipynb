```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
from IPython.display import HTML
from scipy import interpolate
from scipy.optimize import curve_fit 
```

```python
x = np.array([2.3,2.8,3.2,3.7,4.3])
y = np.array([34745,19689,12594,7982,5822])
x_SI = x/100
y_SI = y*1E-6
def func(x,a):
    return a/x**3
fit1 = curve_fit(func,x,y)
x_fit = np.linspace(2.3,4.3,100)
fig = plt.figure()
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
ax.set_title('Campo magnetico de un dipolo a lo largo de su eje')

ax.set_xlabel('x/cm')
ax.set_ylabel(r'B/$\mu$T')
plt.plot(x_fit,func(x_fit,fit1[0][0]))
plt.plot(x,y,"ob")
plt.savefig('momento.png')
```
