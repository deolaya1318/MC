```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
from scipy.fftpack import ifft, fft, fftfreq
data = np.genfromtxt('monthrg.dat')
year = data[:,0]
month = data[:,1]
rg = data[:,3]
year2 = year + month/12
x=year2[3480:4632]
y=rg[3480:4632]
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.set_xlim([1900, 1996])
ax.set_ylim([0, 200])
plt.plot(x,y,"b")
plt.xlabel("t/year")
plt.xlabel("manchas solares/mes")
plt.legend()
plt.show()
```

```python
N = len(y) # Número de puntos de muestreo
dt = x[1]-x[0] # tiempo entre puntos de muestreo
ffty=fft(y)
freq = fftfreq(N, dt) # frecuencias
absffty=np.abs(ffty)
plt.figure(figsize=(10,5))
for i in range(len(absffty)):
    plt.plot([freq[i],freq[i]],[0,absffty[i]],"-")
plt.plot(freq,absffty)
plt.xlabel("f/Hz")
plt.title("DFT")
plt.xlim(-0.5,0.5)
plt.show()
```

```python
freq_max = 0.2
ffty[np.abs(freq) > freq_max] = 0
# Ahora se toma la transformada inversa
y_cleaned=ifft(ffty)
plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x,np.real(y_cleaned),"r", lw = "2",label=u'Señal filtrada')
plt.plot(x,y,"b", lw = "0.3",label=u'Señal original sin ruido')
plt.legend()
plt.xlim([1900, 1996])
plt.ylim([0, 200])
plt.title("frecuencia de corte = "+str(freq_max)+" Hz")
plt.xlabel("t/year")
plt.ylabel("manchas solares/mes")
plt.savefig("fourier.png")
periodo = 1/freq_max
print "El periodo del ciclo solar es:",periodo,"años"
```
