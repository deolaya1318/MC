```python
#Este notebook da respuesta a la tarea 4 del curso
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
from scipy import misc
from scipy.fftpack import ifft, fft, fftfreq, fft2, ifft2, fftn, ifftn, fftshift
```

*Punto a):* Para comenzar se debe realizar la rendija cuadrada de difracción. Usando la herramienta de slice notation podemos asignar a un cuadrado de lado 32 valores distintos a 0 en el centro de un array 2D de lado 1024 lleno de ceros. Luego asignando un mapa de colores se puede graficar la rendija de difracción usando *imshow*.

```python
ladoa = 1024
ladob = 32
a = np.zeros((ladoa,ladoa))
a[(ladoa/2-ladob/2):(ladoa/2+ladob/2),(ladoa/2-ladob/2):(ladoa/2+ladob/2)] = 1
fig, ax = plt.subplots()
im = ax.imshow(a, cmap=plt.get_cmap('hot'))
plt.show()
```

*Punto b):* Ahora del arraylist previamente construido se calcula su DFT y se aplica sobre él la función *fftshift*, para cambiar el orden de las frecuencias. De lo resultante se toma del centro un cuadrado de lado 200 y se produce el patrón de interferencia de esta rendija tomando el cuadrado de la magnitud del vector generado. Por último, se usa *imshow* para gráficar la rendija y el patrón resultante.

```python
plt.figure(figsize=(10,10))
a_dft = fft2(a)
freq = fftshift(a_dft)
ladoc = 200
f = freq[(ladoa/2-ladoc/2):(ladoa/2+ladoc/2),(ladoa/2-ladoc/2):(ladoa/2+ladoc/2)]
fraun = np.log(np.abs(f)**2)
ax = plt.subplot(1,2,1)
ax.imshow(a,interpolation=None,cmap=plt.get_cmap('hot'))
ax = plt.subplot(1,2,2)
ax.set_title("Patron de interferencia de una apertura cuadrada")
ax.imshow(fraun)
plt.show()
```

*Punto c):* Usando el mismo algoritmo que se usaron en los dos previos bloques de código se generó la doble rendija y a partir de las herramientas *fft2* y *fftshift* se encontró su patrón de difracción.

```python
ladoa = 1024
eje1 = 460
eje2 = 564
alto = 200
grosor = 20
a = np.zeros((ladoa,ladoa))
a[(ladoa/2-alto/2):(ladoa/2+alto/2),(eje1-grosor/2):(eje1+grosor/2)] = 1
a[(ladoa/2-alto/2):(ladoa/2+alto/2),(eje2-grosor/2):(eje2+grosor/2)] = 1
plt.figure(figsize=(10,10))
a_dft = fft2(a)
freq = fftshift(a_dft)
ladoc = 128
f = freq[(ladoa/2-ladoc/2):(ladoa/2+ladoc/2),(ladoa/2-ladoc/2):(ladoa/2+ladoc/2)]
fraun = np.abs(f)**2
ax = plt.subplot(1,2,1)
ax.imshow(a,interpolation=None,cmap=plt.get_cmap('hot'))
ax = plt.subplot(1,2,2)
ax.set_title("Patron de difraccion de una doble rendija")
ax.imshow(fraun)
plt.show()
```

*Punto d):* Usando los algoritmos usados previamente se hace una nueva rendija, en este caso con una apertura circular. Para crear esta apertura se realiza un ciclo que recorra todos los indices del lado del cuadrado y evalue la distancia entre el centro del cuadrado y el punto dado por los indices. Si esta distancia es mayor al radio del circulo, el programa devuelve como resultado un cero y en caso contrario devuelve como resultado un 1. Para ello usamos la ecuacion de un circulo y escribimos el ciclo dentro del argumento de la función *array*. Esta solución se inspira en lo encontrado en: http://stackoverflow.com/questions/17163636/filled-circle-in-matrix2d-array. De igual forma que para los anteriores literales se halla el patrón de interferencia creado por esta rendija. Para finalizar se gráfica la interferencia de Fraunhoffer en función de la distancia radial, con el fin de compararla con la figura 8.11 del libro de Born & Wolf, sobre el patrón de interferencia producido por una rendija circular

```python
lado = 1024
radio = 0.01
area=(lado*radio)**2
a = np.array([[1.0*(((x-lado/2)**2+(y-lado/2)**2)<= area) for y in range(lado)] for x in range(lado)])
a_dft = fft2(a)
f = fftshift(a_dft)
fraun = np.log(np.abs(f)**2)
plt.figure(figsize=(10,10))
ax = plt.subplot(1,2,1)
ax.imshow(a,interpolation=None,cmap=plt.get_cmap('hot'))
ax = plt.subplot(1,2,2)
ax.set_title("Patron de interferencia de una apertura circular (Patron de Airy)")
ax.imshow(fraun)
plt.show()
```
